#Básico do Básico - Segmentação de imagens com limiarização

#Threshold: Ultrassom de mama - imagem 1, tumor benígno

###Importando as bibliotecas

```python
%matplotlib inline
import matplotlib.pyplot as plt #para plotar as imagens
import matplotlib.image as mpimg #para importar as imagens
import skimage.io #também para plotar imagens
import numpy as np #para trabalhar com as imagens
#import sys  ?
from skimage.filters import threshold_otsu #função que faz limiarização
from skimage.color import rgb2gray #função que converte imagens em RGB para escala de cinzas
from skimage import exposure #exposição
```

###Importando a imagem

```python
img_mama1 = mpimg.imread('benign (1).png')
```

###Plotando a imagem

```python
imgplot = plt.imshow(img_mama1)
```

###Convertendo a imagem para escala de cinza e definindo o limiar pela função de Otsu

```python
img_mama1_gray = rgb2gray(img_mama1)
thresh = threshold_otsu(img_mama1_gray)
binary = img_mama1_gray > thresh
```

###Imagem final plotada

```python
imgplot = plt.imshow(binary)
```

###Plotando imagem em escala de cinzas e seu histograma

```python
skimage.io.imshow(img_mama1_gray)
histogram, bin_edges = np.histogram(img_mama1_gray, bins=256, range=(0, 1))
plt.figure()
plt.title("Histograma da escala de cinzas")
plt.xlabel("Intensidade do pixel")
plt.ylabel("Número de pixels")
plt.xlim([0.0, 1.0])

plt.plot(bin_edges[0:-1], histogram)
plt.show()
```

###Equalizando a imagem original

```python
img_equalized1 = exposure.equalize_hist(img_mama1)
```

###Plotando duas imagens: a original e sua versão equalizada

```python
imgplot = plt.imshow(img_mama1)
```

```python
imgplot = plt.imshow(img_equalized1)
```

###Com a versão equalizada, transforma-la em escala de cinzas e definindo seu novo limiar

```python
img_mama1_gray_equalized = rgb2gray(img_equalized1)
thresh = threshold_otsu(img_mama1_gray_equalized)
binary4 = img_mama1_gray_equalized > thresh
```

###Plotando duas imagens: a limiarização normal e da imagem equalizada

```python
imgplot = plt.imshow(binary)
```

```python
imgplot = plt.imshow(binary4)
```

###Plotando a imagem equalizada em escala de cinzas e seu histograma

```python
skimage.io.imshow(img_mama1_gray_equalized)
histogram, bin_edges = np.histogram(img_mama1_gray_equalized, bins=256, range=(0, 1))
plt.figure()
plt.title("Histograma da escala de cinzas (equalizada)")
plt.xlabel("Intensidade do pixel")
plt.ylabel("Número de pixels")
plt.xlim([0.0, 1.0])

plt.plot(bin_edges[0:-1], histogram)
plt.show()
```

###Fazendo a equalização adaptativa da imagem original

```python
img_adpt_equalized1 = exposure.equalize_adapthist(img_mama1)
```

###Plotando as duas imagens: a original e a CLAHE

```python
imgplot = plt.imshow(img_mama1)
```

```python
imgplot = plt.imshow(img_adpt_equalized1)
```

###Com a CLAHE, transforma-la em escala de cinzas e definindo seu novo limiar

```python
img_mama1_gray_adpt_equalized = rgb2gray(img_adpt_equalized1)
thresh = threshold_otsu(img_mama1_gray_adpt_equalized)
binary7 = img_mama1_gray_adpt_equalized > thresh
```

###Plotando duas imagens, limirarização normal e a CLAHE

```python
imgplot = plt.imshow(binary)
```

```python
imgplot = plt.imshow(binary7)
```

###Plotando a CLAHE em escala de cinzas e seu histograma

```python
skimage.io.imshow(img_mama1_gray_adpt_equalized)
histogram, bin_edges = np.histogram(img_mama1_gray_adpt_equalized, bins=256, range=(0, 1))
plt.figure()
plt.title("Histograma da escala de cinzas (CLAHE)")
plt.xlabel("Intensidade do pixel")
plt.ylabel("Número de pixels")
plt.xlim([0.0, 1.0])

plt.plot(bin_edges[0:-1], histogram)
plt.show()
```
